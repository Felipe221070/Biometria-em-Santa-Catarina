```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
pd.options.display.max_columns = 60
pd.options.display.max_rows = 90
plt.style.use('classic')
plt.rcParams['figure.figsize'] = (12,8)
```

Pandas é a biblioteca que carrego na memória com um apelido pd, matplotlib permite visualizações básicas e faremos os gráficos diretamente nos dataframes

```python
# O arquivo lido é atribuido a variável 'filename1'.
filename1 = pd.read_csv(r'C:\Users\Hariel\perfil_eleitor_secao_ATUAL_SC.csv', sep = ";", error_bad_lines=False, encoding= "Latin 1")
```

```python
filename1.info()
```

'NM_MUNICIPIO' = Nome do minucipio onde vota o eleitor

'CD_MUNICIPIO' = Código TSE do município onde vota o eleitor

'QT_ELEITORES_PERFIL' = Quantitativo de eleitores aptos a votar

'QT_ELEITORES_BIOMETRIA' = Quantitativo de eleitores aptos, com biometria cadastrada

'CD_MUN_SIT_BIOMETRIA' = Código da situação biométrica do município, '-1' (quando a eleição é totalmente biométrica), '-2' (quando a eleição pode ser com ou sem biométria) e '-3' (quando a eleição é sem biometria)


```python
filename1.sample(10)
```

```python
filename1.shape
```

```python
filename1.groupby('NM_MUNICIPIO').size().sort_values().tail(10).plot(kind = 'barh', title = 'Nome do município onde vota o eleitor', ylabel = ' ')
```

```python
filename1.groupby('CD_MUNICIPIO').size().sort_values().tail(10).plot(kind = 'barh', title = 'Número do TSE do município onde vota o eleitor', ylabel = ' ')
```

```python
filename1.groupby('QT_ELEITORES_PERFIL').size().sort_values().tail(10).plot(kind = 'barh', title='Quantitativo de eleitores aptos a votar por município')
```

```python
filename1.groupby('QT_ELEITORES_BIOMETRIA').size().sort_values().tail(10).plot(kind = 'barh', title='Quantitativo de eleitores aptos, com biometria cadastrada por município')
```

```python
# -1 = Quando a eleição é totalmente biométrica,
# -2 = Quando a eleição pode ser com ou sem biometria,
# -3 = Quando a eleição é sem biometria.

filename1.CD_MUN_SIT_BIOMETRIA = filename1.CD_MUN_SIT_BIOMETRIA.astype(str)
filename1['CIDADE_BIOMETRIA:'] = filename1['NM_MUNICIPIO'] + ' - Código da situação biométrica do município = ' + filename1['CD_MUN_SIT_BIOMETRIA'] + '       ' + ' Número total de eleitores = '
filename1.groupby('CIDADE_BIOMETRIA:').size().sort_values().tail(7)
```
